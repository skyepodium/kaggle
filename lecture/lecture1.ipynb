```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import math
```

```python
CauseOfDeath = [
    "Heart disease",
    "Cancer",
    "Cerebrovascular",
    "Chronic respiratory",
    "Accidents",
    "Diabetes mellitus",
    "Flu and pneumonia",
    "Alzheimer's disease",
    "Kidney disorders",
    "Septicemia"
]

Counts = [
    700142,
    553768,
    163538,
    123013,
    101537,
    71372,
    62034,
    53852,
    39480,
    32238
]
```

```python
data = pd.DataFrame({
    "CauseOfDeath": CauseOfDeath,
    "Counts": Counts
})
```

```python
data["Counts2"] = data["Counts"] / 1000
data["Counts2"] = data["Counts2"].apply(math.trunc)
```

```python
data.head()
```

```python
color = sns.color_palette("Set1", 10)
sns.palplot(color)
```

```python
# figure, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots()
# figure.set_size_inches(18, 8)
sns.set_style("whitegrid")
plt.figure(figsize=(14, 8))
plt.xticks(rotation=40, ha="right")
sns.set(font_scale=2)
sns.set_palette(color)
ax = sns.barplot(data=data, x="CauseOfDeath", y="Counts2")
ax.set(xlabel="Cause Of Death", ylabel="Counts(x1000)")
plt.show()
```
