```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv("./input/train.csv")
test = pd.read_csv("./input/test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
def setEmbarked(val):
    result = None
    if val == "S": result = 2
    elif val == "C": result = 1
    else: result = 0
    return result
        
```

```python
train["Embarked"] = train["Embarked"].apply(setEmbarked)
test["Embarked"] = test["Embarked"].apply(setEmbarked)
```

```python
def setSex(val):
    result = None
    if val == "male": result = 1
    else: result = 0
    return result
```

```python
train["Sex"] = train["Sex"].apply(setSex)
test["Sex"] = test["Sex"].apply(setSex)
```

```python
train.head()
```

```python
def setCabin(val):
    result = 0
    if pd.isna(val): result = 0
    else: result = 1
    return result
```

```python
train["HasCabin"] = train["Cabin"].apply(setCabin)
test["HasCabin"] = test["Cabin"].apply(setCabin)
train = train.drop(columns=["Cabin"], axis=1)
test = test.drop(columns="Cabin", axis=1)
```

```python
test["Fare"] = test["Fare"].fillna(test["Fare"].median())
```

```python
train["FamilySize"] = train["SibSp"] + train["Parch"]
test["FamilySize"] = test["SibSp"] + test["Parch"]
train = train.drop(columns=["SibSp", "Parch"], axis=1)
test = test.drop(columns=["SibSp", "Parch"], axis=1)
```

```python
for item in train["Name"]:
    print(item)
```

```python
train["Title"] = train["Name"].str.extract(', ([A-Z][a-z]+)')
test["Title"] = test["Name"].str.extract(', ([A-Z][a-z]+)')
```

```python
train["Title"].value_counts()
```

```python
def setTitle(val):
    result = 0
    if val == "Mr": result = 0
    elif val == "Miss": result = 1
    elif val == "Mrs": result = 2
    elif val == "Master": result = 3
    else: val = 4
    return result
```

```python
train["Title"] = train["Title"].apply(setTitle)
test["Title"] = test["Title"].apply(setTitle)
```

```python
train = train.drop(columns=["Ticket", "Name"], axis=1)
test = test.drop(columns=["Ticket", "Name"], axis=1)
```

```python
box = train.boxplot(column=["Fare"])
```

```python
train["FareBand"] = pd.cut(train["Age"], 5)
train[["FareBand", "Survived"]].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
```

```python
def setFare(val):
    result = 0
    if val < 16.336: result = 0
    elif val < 32.252: result = 1
    elif val < 48.168: result = 2
    elif val < 64.084: result = 3
    else: result = 4
    return result
```

```python
train["Fare"] = train["Fare"].apply(setFare)
test["Fare"] = test["Fare"].apply(setFare)
```

```python
train.corr()
```

```python
train_age_yes = train[train["Age"].notna()]
train_age_no = train[train["Age"].isna()]
```

```python
x = train_age_yes[["Pclass", "Fare", "HasCabin", "FamilySize", "Title"]]
y = train_age_yes["Age"]
lr = LinearRegression()
lr.fit(x, y)
```

```python
x_test = train_age_no[["Pclass", "Fare", "HasCabin", "FamilySize", "Title"]]
y_predict = lr.predict(x_test)
train_age_no["Age"] = y_predict
```

```python
train = pd.concat([train_age_yes, train_age_no])
```

```python
train["Age"] = train["Age"].round()
```

```python
test_age_yes = test[test["Age"].notna()]
test_age_no = test[test["Age"].isna()]
```

```python
x = test_age_yes[["Pclass", "HasCabin", "Title"]]
y = test_age_yes["Age"]
lr.fit(x, y)
```

```python
x_test = test_age_no[["Pclass", "HasCabin", "Title"]]
y_predict = lr.predict(x_test)
test_age_no["Age"] = y_predict
```

```python
test = pd.concat([test_age_yes, test_age_no])
```

```python
test["Age"] = test["Age"].round()
```

```python
train
```

```python
train["AgeBand"] = pd.cut(train["Age"], 5)
train[["AgeBand", "Survived"]].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
def setAge(val):
    if val < 16: result = 0
    elif val < 32: result = 1
    elif val < 48: result = 2
    elif val < 64: result = 3
    else: result = 4
    return result
```

```python
train["Age"] = train["Age"].apply(setAge)
test["Age"] = test["Age"].apply(setAge)
```

```python
def setAlone(val):
    if val == 0 : return 1
    else: return 0
```

```python
train["IsAlone"] = train["FamilySize"].apply(setAlone)
test["IsAlone"] = test["FamilySize"].apply(setAlone)
```

```python
x = train[["Pclass", "Sex", "Age", "Fare", "Embarked", "HasCabin", "FamilySize", "Title", "IsAlone"]]
y = train["Survived"]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=30, shuffle=True, random_state=0)

scoring = 'accuracy'
```

```python
rf = RandomForestClassifier()
rf.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(rf, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
print("RandomForest: {0: .4f}".format(round(np.mean(score)*100, 2)))
```

```python
lo = LogisticRegression()
lo.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(lo, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
print("LogisticRegression: {0: .4f}".format(round(np.mean(score)*100, 2)))
```

```python
sv = SVC()
sv.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(sv, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
print("SupportVectorMachine: {0: .4f}".format(round(np.mean(score)*100, 2)))
```

```python
kn = KNeighborsClassifier()
kn.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(kn, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
print("KNeighborsClassifier: {0: .4f}".format(round(np.mean(score)*100, 2)))
```

```python
ga = GaussianNB()
ga.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(ga, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
print("GaussianNB: {0: .4f}".format(round(np.mean(score)*100, 2)))
```

```python
pc = Perceptron()
pc.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(pc, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
print("perceptron: {0: .4f}".format(round(np.mean(score)*100, 2)))
```

```python
sgd = SGDClassifier()
sgd.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(sgd, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
print("SGD: {0: .4f}".format(round(np.mean(score)*100, 2)))
```

```python
de = DecisionTreeClassifier()
de.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(de, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
print("DecisionTree: {0: .4f}".format(round(np.mean(score)*100, 2)))
```

```python
xgb_model = xgb.XGBClassifier(objective="binary:logistic", random_state=42)
xgb_model.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(xgb_model, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
print("xgboost: {0: .4f}".format(round(np.mean(score)*100, 2)))
```

```python
x_test = test[["Pclass", "Sex", "Age", "Fare", "Embarked", "HasCabin", "FamilySize", "Title", "IsAlone"]]
```

```python
predict = xgb_model.predict(x_test)
```

```python
submission = pd.DataFrame({
    "PassengerId": test["PassengerId"],
    "Survived": predict
})
```
