```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("./input/train.csv")
test = pd.read_csv("./input/test.csv")

train_test = [ train, test ]
```

```python
### 2) check Missing Value ( NaN )
```

```python
name_list = ["train", "test"]
def check_na():
     for (index, item) in enumerate(train_test):
        print("[ {} NaN List ]".format(name_list[index]))
        print(item.isna().sum())
        print("\n")
check_na()
```

```python
# 3. NaN
```

```python
### 1) Embarked - train
```

```python
train["Embarked"] = train["Embarked"].fillna(train["Embarked"].mode()[0])
train["Embarked"].isna().sum()
```

```python
### 2) Fare - test
```

```python
test["Fare"] = test["Fare"].fillna(test["Fare"].median())
train["Fare"].isna().sum()
```

```python
### 3) Cabin - train, test
```

```python
def set_cabin(val):
    return 1 if not pd.isna(val) else 0
```

```python
for item in train_test:
    item["HasCabin"] = item["Cabin"].apply(set_cabin)
    item.drop(columns=["Cabin"], inplace=True)
```

```python
train.head(1)
```

```python
test.head(1)
```

```python
### 4) Age - train, test
```

```python
train.corr()
```

```python
train["Pclass"].value_counts()
```

```python
age_list = train.groupby("Pclass")["Age"].median() 
print(age_list)
```

```python
def set_age(val):
    Pclass = val["Pclass"]
    Age = val["Age"]
    
    result = Age
    if pd.isna(Age):
        return age_list[Pclass]
    else:
        return result
```

```python
for item in train_test:
    item["Age"] = item.apply(set_age, axis=1)
```

```python
check_na()
```

```python
# 3. EDA
# Ticket, Name, SibSp, Parch
```

```python
### 1) Ticket
```

```python
train["Ticket"].value_counts()
```

```python
for item in train_test:
    item.drop(columns=["Ticket"], inplace=True)
```

```python
### 2) Name
```

```python
train["Name"].value_counts()
```

```python
for item in train_test:
    item["Title"] = item["Name"].str.extract(', ([A-Z][a-z]+)')
```

```python
train["Title"].value_counts()
```

```python
# https://en.wikipedia.org/wiki/Master_(form_of_address)
```

```python
def set_title(val):
    result = ""
    if val == "Mr" or val =="Miss" or val == "Mrs" or val == "Master": result = val
    else: result = "Others"
    return result

for item in train_test:
    item["Title"] = item["Title"].apply(set_title)
    item.drop(columns=["Name"], inplace=True)
```

```python
train["Title"].value_counts()
```

```python
### 3) SibSp, Parch
```

```python
for item in train_test:
    item["FamilySize"] = item["SibSp"] + item["Parch"]
```

```python
def set_is_alone(val):
    result = 1
    if val == 0: result =1
    else: result = 0
    return result

for item in train_test:
    item["IsAlone"] = item["FamilySize"].apply(set_is_alone)
```

```python
train.head()
```

```python
# 4. Binning
# Age, Fare
```

```python
train["AgeBand"] = pd.cut(train["Age"], 4)
train[["AgeBand", "Survived"]].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
def set_age(val):
    if val < 20.315: result = 3
    elif val < 40.21: result = 2
    elif val < 60.105: result = 1
    else: result = 0
    return result

for item in train_test:
    item["Age"] = item["Age"].apply(set_age)

train = train.drop(columns=["AgeBand"], axis=1)
```

```python
### 2) Fare
```

```python
train["FareBand"] = pd.cut(train["Fare"], 4)
train[["FareBand", "Survived"]].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
```

```python
def set_fare(val):
    result = 0
    if val < 128.082: result = 3
    elif val < 256.165: result = 2
    elif val < 384.247: result = 1
    else: result = 0
    return result
```

```python
train["Fare"] = train["Fare"].apply(set_fare)
test["Fare"] = test["Fare"].apply(set_fare)
train.drop(columns=["FareBand"], axis=1, inplace=True)
```

```python
# 5. one_hot_encoding
# sex, Fare, Embarked, HasCabin, Title, IsAlone
```

```python
train.head(1)
```

```python
encoding_list = ["Sex", "Fare", "Embarked", "HasCabin", "Title"]
```

```python
train = pd.get_dummies(train, columns=encoding_list, drop_first=True)
test = pd.get_dummies(test, columns=encoding_list, drop_first=True)
```
