```python
def setEmbarked(val):
    result = None
    if val == "S": result = 2
    elif val == "C": result = 1
    else: result = 0
    return result
```

```python
train["Embarked"] = train["Embarked"].apply(setEmbarked)
test["Embarked"] = test["Embarked"].apply(setEmbarked)
```

```python
def setAge(val):
    if val < 16: result = 0
    elif val < 32: result = 1
    elif val < 48: result = 2
    elif val < 64: result = 3
    else: result = 4
    return result
```
