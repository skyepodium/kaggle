```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
ml = LogisticRegression(solver='lbfgs')
x = x_table
y = y_table
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

scoring = 'accuracy'
score = cross_val_score(ml, x_table, y_table, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)

round(np.mean(score)*100, 2)
```

```python
lf = RandomForestClassifier()
lf.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(lf, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)

round(np.mean(score)*100, 2)
```

```python
lf = KNeighborsClassifier()
lf.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(lf, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)

round(np.mean(score)*100, 2)
```

```python
lf = DecisionTreeClassifier()
lf.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(lf, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)

round(np.mean(score)*100, 2)
```

```python
lf = GaussianNB()
lf.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(lf, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)

round(np.mean(score)*100, 2)
```

```python
lf = SVC()
lf.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(lf, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)

round(np.mean(score)*100, 2)
```

```python
lf = DecisionTreeClassifier()
lf.fit(x, y)
scoring = 'accuracy'
score = cross_val_score(lf, x, y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)

round(np.mean(score)*100, 2)
```
